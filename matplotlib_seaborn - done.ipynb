#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.

* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
monthly_steps = fitbit.groupby('Months_encoded')['Steps'].median()
fig = plt.figure(figsize=(12,4))
plt.bar(monthly_steps.index, monthly_steps.values, color='green')
plt.xticks(monthly_steps.index, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'Jul','Aug','Sep','Oct','Nov','Dec'])

plt.savefig('figures/monthly_steps_mean.png', dpi=600)
plt.show()




# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
ww_steps = fitbit.groupby('Work_or_Weekend')['Steps'].median()
fig2=plt.figure(figsize=(12,4))
plt.barh(ww_steps.index, ww_steps.values, color='yellow')
plt.yticks([0,1],['Weekend', 'Workday'])
plt.show()
plt.savefig('figures/ww_median.png', dpi=600)
```

#### Use the right visulalization to show the distribution of column `Age`.

```python
# your code here
# Plot histogram
plt.hist(titanic['Age'], bins=20, edgecolor='black')

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')

plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
# your code here
bin_sizes = [10, 20, 50]

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=len(bin_sizes), figsize=(15, 5))

# Plot histograms for each bin size
for i, bins in enumerate(bin_sizes):
    axes[i].hist(titanic['Age'], bins=bins, edgecolor='black')
    axes[i].set_xlabel('Age')
    axes[i].set_ylabel('Frequency')
    axes[i].set_title(f'Distribution of Age (Bins={bins})')

# Adjust the spacing between subplots
plt.tight_layout()

plt.show()
```
