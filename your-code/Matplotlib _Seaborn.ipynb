#### Use the right visulalization to show the distribution of column `Age`.

```python
# your code here
import pandas as pd
import matplotlib.pyplot as plt

# Set the plot style to classic
plt.style.use('classic')

# Read the Titanic dataset
file_name = 'C:/Users/hilu1/ironhack/week4/lab-matplotlib-seaborn/your-code/titanic.csv'
titanic = pd.read_csv(file_name)

# Set the figure size
plt.figure(figsize=(12, 6))

# Drop missing values from the 'Age' column
ages = titanic['Age'].dropna()

# Create the histogram
plt.hist(ages, bins=30, edgecolor='k', alpha=0.7)

# Add titles and labels
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Add grid lines
plt.grid(True)

# Tighter layout
plt.tight_layout()

# Show the plot
plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
# your code here
# Drop missing values from the 'Age' column
ages = titanic['Age'].dropna()

# Set up the figure and axes for subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

# Plot for bin size = 10
axes[0].hist(ages, bins=10, edgecolor='k', alpha=0.7)
axes[0].set_title('Histogram with Bins = 10')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')

# Plot for bin size = 20
axes[1].hist(ages, bins=20, edgecolor='k', alpha=0.7)
axes[1].set_title('Histogram with Bins = 20')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Frequency')

# Plot for bin size = 50
axes[2].hist(ages, bins=50, edgecolor='k', alpha=0.7)
axes[2].set_title('Histogram with Bins = 50')
axes[2].set_xlabel('Age')
axes[2].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

#### How does the bin size affect your plot?

```python

Observations Based on Different Bin Sizes
Bin Size = 10: This results in fewer bins and wider intervals, giving a general sense of the age distribution with less 
detail. It may smooth out the distribution and show broad trends.

Bin Size = 20: With more bins, the histogram shows more detail and can reveal finer variations in the age distribution. 
Peaks and modes in the distribution might become more apparent.

Bin Size = 50: Even more bins provide a highly detailed view of the age distribution, potentially revealing smaller peaks
or variations that were not as visible with fewer bins. This can provide a more nuanced understanding of the distribution's shape and characteristics.

Conclusion
Choosing an appropriate bin size depends on the nature of the data and the insights you want to extract. Too few bins can
oversimplify the distribution, while too many bins can obscure patterns with noise. Experimenting with different bin sizes
helps to find a balance between visual clarity and informative detail in histogram visualizations. Adjusting bin sizes
allows you to tailor the histogram to best reveal the underlying distribution of the data, providing insights into the
dataset's characteristics.
```

#### Plot the correlation matrix using seaborn.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the Titanic dataset (adjust the file path accordingly)
file_name = 'C:/Users/hilu1/ironhack/week4/lab-matplotlib-seaborn/your-code/titanic.csv'
titanic = pd.read_csv(file_name)

# Select numerical columns for correlation matrix
numerical_cols = titanic.select_dtypes(include=['float64', 'int64']).columns

# Calculate correlation matrix
corr_matrix = titanic[numerical_cols].corr()

# Plot correlation matrix using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True, fmt='.2f', linewidths=.5)

# Adding title
plt.title('Correlation Matrix of Numerical Features')

# Show plot
plt.show()
```
